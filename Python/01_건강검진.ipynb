```python
df2.columns = ['성별코드', '허리둘레', '수축기 혈압', '이완기 혈압', '식전혈당(공복혈당)', '트리글리세라이드',
       'HDL 콜레스테롤']
       #여기서 고치면 열이름 수정된다. or 밑에 줄
```

```python
# 오브젝트인 df2.columns를 list형태로 받음
lt = list(df2.columns)
lt
```

```python
# 공백제거 한번에 해준다
lt = [c.replace(' ','') for c in lt]
lt
```

```python
# 부분만 고치기
lt[4] = '공복혈당'
lt
```

```python
# 바뀐 값을 다시 반영
df2.columns=lt
df2.info()
```

```python
# 빈칸 제거
df2 = df2.dropna(axis=0)
df2.info()
```

```python
import numpy as np
```

+ 분석해봄

```python
df2['높은혈압'] = np.where((df2['수축기혈압'] >= 130) | (df2['이완기혈압'] >= 85), 1, 0)
df2.head()
```

```python
df2['높은혈당'] = np.where((df2['공복혈당'] >= 100), 1, 0)
df2.head()
```

```python
df2['높은중성지방'] = np.where((df2['트리글리세라이드'] >= 150), 1, 0)
df2.head()
```

```python
df2['낮은HDL콜레스테롤수치'] = np.where(((df2['성별코드'] == 1) & (df2['HDL콜레스테롤']<40)|(df2['성별코드'] == 2)& (df2['HDL콜레스테롤']<50)), 1, 0)
df2.head()
```

```python
df2['복부비만'] = np.where(((df2['성별코드'] == 1) & (df2['허리둘레'] >= 90)|(df2['성별코드'] == 2)& (df2['허리둘레'] >= 85)), 1, 0)
df2.head()
```

```python
df2['대사증후군'] = (df2['높은혈압']+
                    df2['높은혈당']+
                    df2['높은중성지방']+
                    df2['낮은HDL콜레스테롤수치']+
                    df2['복부비만'])
```

```python
df2['판별'] = np.where((df2['대사증후군'] == 0), "정상",
np.where((df2['대사증후군'] <= 2), "주의군", "위험군"))
```

```python
df2['성별'] = np.where(df2['성별코드'] == 1,'남','여')
```

```python
df3=df2.groupby(['성별', '판별']).count()[['성별코드']]
df3
```
