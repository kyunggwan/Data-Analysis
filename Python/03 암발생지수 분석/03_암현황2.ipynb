```python
import numpy as np
```

```python
df['발생자수'] = np.where(df['발생자수'] == '-', 0, df['발생자수'])
```

```python
df['조발생률'] = np.where(df['조발생률'] == '-', 0, df['조발생률'])
```

```python
df.info()
```

```python
# 데이터 타입 변경
# 오브젝트 형을 numeric으로
df['발생자수'] = df['발생자수'].astype('int')
df.info()
```

```python
# 데이터 타입 변경
df['조발생률'] = pd.to_numeric(df['조발생률'])
df.info()
```

```python
# 암 종류
lt = list(df['암종별'].unique())
lt
```

```python
lt = [ s[-3:-1] for s in lt]
#숫자 부분만 출력했지만 수치형은 아님
lt
```

```python
# int 형으로 변경
lt = [int(s) for s in lt]
lt
```

```python
temp = list(df['발생자수'])
df.info()
```

```python
temp = [int(s.replace('-', '0')) for s in temp]
df['발생자수'] = temp
df.info()
```

```python
temp = list(df['조발생률'])
temp = [float(s.replace('-', '0')) for s in temp]
df['조발생률'] = temp
df.info()
```

```python
def func(col, df):
    temp = list(df[col])
    temp = [float(s.replace("-",'0')) for s in temp]
    df[col] = temp
```

```python
for c in ['발생자수', '조발생률']:
    func(c, df)
```
